```
from multiple_choice import generate_multiple_choice
from tools import MultipleChoiceArgs

args = MultipleChoiceArgs()
args.topic = "Амур"
args.max_questions = 10
args.generate_count = 10

questions = generate_multiple_choice(args)
```

```
import ipywidgets as widgets
import random 

empty_value='Не знаю'
objects = []
radios = {}
labels = {}
for q in questions:
  label = widgets.Label(value = f"{q['question']}")
  ans = [q['answer']] + q['wrong']
  random.shuffle(ans)
  radio = widgets.RadioButtons(options=[empty_value] + ans,layout={'width': 'max-content'},value=empty_value)
  objects.extend([label, radio])
  radios[q['question']] = radio
  labels[q['question']] = label

button = widgets.Button(description='Проверка')

user_answers = {}

def select_data(**kwargs):
    for key in kwargs:
        user_answers[key] = kwargs[key]
            
def on_button_clicked(b):
    n_correct = 0
    for index, q in enumerate(questions):
      labels[q['question']].value = ("✅" if (user_answers[q['question']] == q['answer']) else "❌") + " " + q['question']
      n_correct += 1 if (user_answers[q['question']] == q['answer']) else 0
    print(f'Верных ответов {n_correct} из {len(questions)}')

button.on_click(on_button_clicked)

out = widgets.interactive_output(select_data, radios)
ui = widgets.VBox(children=objects+[button])
display(ui, out)

```
